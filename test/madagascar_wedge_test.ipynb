```python
#m8r example after Sergey Fomel
```

```python
top = m8r.math(n1=301,d1=1,o1=0,label1='Trace',output=0.06)[0]
bot = m8r.math(output='input*(2-x1/300)')[top]

wedge = m8r.cat(axis=2).unif2(n1=181,o1=0,d1=0.001,unit1='s',label1='Time',v00=[10,15,20])[top,bot]
seis = m8r.ai2refl.ricker1(frequency=25)[wedge]

seis.grey(title='Wedge',color='G')
wind = m8r.window(j2=10)[seis]
wind.wiggle(title='WEDGE',poly=True,yreverse=True,transp=True)
```
